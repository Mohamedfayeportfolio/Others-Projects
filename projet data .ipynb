```python
#Creation d'une colonnes tranche d'age entre 18ans et 25 ans, de 25 a 50 et plus : Creation of an age range column from 18 to 25, from 25 to 50 and over 
#np.nan
import numpy as np
df['Age_range'] = np.nan
```

```python
df.columns
```

```python
df['Age_range'][df['age']< 18] = "<18"
df['Age_range'][(df['age']>= 18) & (df['age'] < 25)] = "18-25"
df['Age_range'][(df['age']>= 25) & (df['age'] < 50)] = "25-50"
df['Age_range'][(df['age']>= 50)] = ">=50"
```

```python
#le solde moyen des individu par tranches d'age
#average balance of individuals by age group
df.groupby("Age_range")["balance"].mean()
```

```python
#filtre des clients de plus de 25 ans et calcule l'age moyen en fonction du niveau d'Ã©ducation 
#filter customers over 25 and calculate average age based on education level 
df[df["age"]>25].groupby('education')["age"].mean()
```

```python
# Segmentation par profession et statut marital
segmentation = df.groupby(['job', 'marital']).agg({
    'age': 'mean',              
    'balance': 'mean',          
    'duration': 'mean',                          
}).reset_index()
```

```python
# Ajouter une colonne pour le nombre total de clients dans chaque segment
segmentation['total_clients'] = df.groupby(['job', 'marital']).size().reset_index(name='total_clients')['total_clients']
```

```python
# Trier par nombre total de clients (pour identifier les segments les plus importants)
segmentation = segmentation.sort_values(by='total_clients', ascending=False)
```

```python
# Afficher les rÃ©sultats
segmentation
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Tracer un graphique en barres pour les segments les plus importants
plt.figure(figsize=(14, 8))
sns.barplot(data=segmentation.head(10), x='job', y='total_clients', hue='marital')
plt.title('Top 10 des Segments par Profession et Statut Marital')
plt.xlabel('Profession')
plt.ylabel('Nombre de Clients')
plt.xticks(rotation=45)
plt.legend(title='Statut Marital')
plt.show()
```
